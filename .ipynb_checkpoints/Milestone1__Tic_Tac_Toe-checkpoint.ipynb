```python
def display_board(board):
    clear_output()
    
    print('   |   |   ')
    print(' ' + board[7] +' | ' + board[8] + ' | ' + board[9] + ' ')
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' ' + board[4] +' | ' + board[5] + ' | ' + board[6] + ' ')
    print('   |   |   ')
    print('-----------')
    print('   |   |   ')
    print(' ' + board[1] +' | ' + board[2] + ' | ' + board[3] + ' ')
    print('   |   |   ')
```

```python
def player_in(count, p1, p2, pos_init, board):
    legal_in = [1,2,3,4,5,6,7,8,9,'1','2','3','4','5','6','7','8','9']
    
    # If count is odd, then it is p1's turn
    if count%2==1:
        player = p1
    else:  # else, it's p2's turn
        player = p2
    
    pos=input('Choose your position: (1-9)')
    while pos not in legal_in:
        print('Please enter again')
        pos=input('Choose your position: (1-9)')
    pos = int(pos)
    
    # If the position is empty, then it is able to mark
    if board[pos] == ' ':
        count += 1
        board[pos] = player
        display_board(board)
    # If the position is not empty, it is not able to mark
    else:
        print('This position has been chosen')
    return count
```

```python
# Determine if one of the player wins the game
def end_cond(board):
    if board[1]==board[4]==board[7]=='X' or board[1]==board[4]==board[7]=='O':
        return True
    elif board[2]==board[5]==board[8]=='X' or board[2]==board[5]==board[8]=='O':
        return True
    elif board[3]==board[6]==board[9]=='X' or board[3]==board[6]==board[9]=='O':
        return True
    elif board[1]==board[2]==board[3]=='X' or board[1]==board[2]==board[3]=='O':
        return True
    elif board[4]==board[5]==board[6]=='X' or board[4]==board[5]==board[6]=='O':
        return True
    elif board[7]==board[8]==board[9]=='X' or board[7]==board[8]==board[9]=='O':
        return True
    elif board[1]==board[5]==board[9]=='X' or board[1]==board[5]==board[9]=='O':
        return True
    elif board[3]==board[5]==board[7]=='X' or board[3]==board[5]==board[7]=='O':
        return True
    else:
        return False
```

```python
from IPython.display import clear_output
while(True):
    clear_output()
    print('Welcome to Tic Tac Toe')
    end_flag = False  # The flag to determine if the game is over
    count = 1  # Whose turn
    while(True):
        p1 = input('Do you want to be X or O? ')
        p1 = p1.upper()
        if p1=='X' or p1=='O':
            break
    if p1=='X':
        p2 = 'O'
    else:
        p2 = 'X'
    print('Player 1 will go first.')
    while (True):
        play_flag = input('\nAre you ready to play? Enter Yes or No. ')
        play_flag = play_flag.lower()
        if play_flag=='yes' or play_flag=='no':
            break
    if play_flag != 'yes':
        clear_output()
        print('Game Over')
        break
    
    clear_output()
    
    # initial the board
    board = [' ']*10
    display_board(board)
    
    while(not end_flag and count!=10):
        count = player_in(count, p1, p2, pos_init, board)
        end_flag = end_cond(board)
    
    # If the board is full, then no one wins the game
    if count==10:
        print('The game is over, no one wins the game!')
    # If someone wins the game
    else:
        print('Congratulations! You have won the game!')
    again = input('Do you want to play again? Enter Yes or No : ')
    again = again.lower()
    if again!='yes' or again!='no':
        print('Please enter again')
        again = input('Do you want to play again? Enter Yes or No : ')
    if again=='no':
        break
```
